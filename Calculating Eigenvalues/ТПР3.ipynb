Запрос количества альтернатив

```python
num = 4#@param
```

Создание матрицы

```python
arr = np.ones((num, num))
arr
```

Матрица парных сравнений

```python
for i in range(0, num):
  for j in range(i+1, num):
    print(f"Как {i+1} относится к {j+1}?")
    relation = np.float32(input())
    arr[i, j] = relation
    arr[j, i] = 1/relation
pd.DataFrame(arr, index=range(1, num+1), columns=range(1, num+1))
```

Ввод погрешности

```python
error = 0.00001#@param
```

Итерации

```python
Xarr = np.zeros(num)
Rarr = np.zeros(num)
Xarr[0] = 1
k = 0
S = 0
Xarr
```

```python
while(True):
  k +=1
  for i in range(0, num):
    Rarr[i] = 0
    for j in range(0, num):
      Rarr[i] = Rarr[i] + arr[i, j] * Xarr[j]
  for i in range(0, num):
    Xarr[i] = Rarr[i]
  L = Xarr[0]
  for i in range(1, num):
    if Xarr[i] > L:
      L = Xarr[i]
  for i in range(0, num):
    Xarr[i] = Xarr[i]/L
  print(f"k = {k}, L = {L}, x = {list(Xarr)}")
  if ~(abs(L-S) > error):
    break
  S = L

print("Результат вычисления:")
print(f'Наибольшее собственное значение L = {L}');
print(f'Собст. вектор X: {list(Xarr)}');
print(f'Число итераций k = {k}');
```
