```python
import numpy as np
```

Исходные данные:

```python
arr = np.array([[180, 176, 176, 181, 177, 180], [67, 70, 68, 67, 68, 66], [850, 1002, 860, 818, 860, 801]], dtype=float).transpose()
arr
```

Вес каждого параметра:

```python
weight = np.array([0.4, 0.24, 0.36])
weight
```

Если True - для параметра оптимальна максимизация, если False - минимизация

```python
maximize = np.array([True, True, False])
```

Поиск максимума:

```python
max_crit = arr.max(0)
max_crit
```

Нормализация критериев:

```python
arr = arr/max_crit
arr
```

```python
for n in range(0, len(maximize)):
  if ~maximize[n]:
    arr.transpose()[n] = 1 - arr.transpose()[n] 
arr
```

Поиск обобщённой функции:

```python
arr *= weight
arr
```

```python
func = arr.sum(1)
func
```

```python
func.max()
```

Результатом будет номер оптимального двигателя.

```python
result = list(func).index(func.max()) + 1
result
```
